```python
from agents import Agent, Runner, set_default_openai_client, set_tracing_disabled, HostedMCPTool
from agents.models.openai_provider import OpenAIProvider
from agents.run import RunConfig
from openai import AsyncOpenAI
import os
```

```python
client = AsyncOpenAI(base_url="http://localhost:8321/v1/openai/v1", api_key="fake")
provider = OpenAIProvider(openai_client=client)
```

```python
set_tracing_disabled(disabled=True)
```

```python

agent = Agent(
    name="KubernetesAgent",
    model="Qwen/Qwen3-0.6B",
    tools=[
        HostedMCPTool(
            tool_config={
                "type": "mcp",
                "server_label": "kubernetes",
                "server_url": "http://localhost:8080/sse",
                "require_approval": "never",
            },
        )
    ],
)
```

```python
result = await Runner.run(
    agent,
    "Using the tool provided,list all pods in the kube-system namespace",
    run_config=RunConfig(model_provider=provider),
)
print(result.final_output)
```
