```python
from openai import OpenAI
```

```python
model = "Qwen/Qwen3-0.6B"
# Initialize OpenAI client
client = OpenAI(base_url="http://localhost:8321/v1", api_key="fake")
```

```python
tools=[
        {
            "type": "mcp",
            "server_label": "kubernetes",
            "server_url": "http://localhost:8080/mcp",
            "require_approval": "always",
        },
    ]
```

```python
response = client.responses.create(
            model=model,
            input="Using the tool provided, list all pods in the kube-system namespace",
            tools=tools,
        )
print(response)
```

```python
response = client.responses.create(
            model=model,
            input=[
                {
                    "type": "mcp_approval_response",
                    "approve": True,
                    "approval_request_id": "approval_37512293-2603-46a9-b8c7-b6aa60e96029"
                }
            ],
            tools=tools,
            previous_response_id="resp-6c4f7654-68df-47c7-83dc-df7a1cc737cb"
        )
print(response.output_text)
```
